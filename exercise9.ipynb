```python
import pandas as pd
import seaborn as sns
import numpy as np

from sklearn import metrics

from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.model_selection import cross_val_score, KFold

from sklearn import datasets
```

## Dataset File

```python
dataset_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/train.csv?raw=true'
df = pd.read_csv(dataset_url)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df['person_home_ownership'].value_counts()
```

```python
df['person_home_ownership'] = df['person_home_ownership'].map({'RENT': 0, 'MORTGAGE': 1, 'OWN':2, 'OTHER':3})
df.head()
```

```python
df['loan_intent'].value_counts()
```

```python
df['loan_intent'] = df['loan_intent'].map({'EDUCATION': 0, 'MEDICAL': 1, 'PERSONAL':2, 'VENTURE':3, 'DEBTCONSOLIDATION':4, 'HOMEIMPROVEMENT':5})
df.head()
```

```python
df['loan_grade'].value_counts()
```

```python
df['loan_grade'] = df['loan_grade'].map({'A': 0, 'B': 1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6})
df.head()
```

```python
df['cb_person_default_on_file'].value_counts()
```

```python
df['cb_person_default_on_file'] = df['cb_person_default_on_file'].map({'N': 0, 'Y': 1})
df.head()
```

```python
df.drop(['id'], axis=1, inplace=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop("loan_status",axis=1)
y = df["loan_status"]

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.3)
```

## Test File

```python
test_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/test.csv?raw=true'
dt=pd.read_csv(test_url)
```

```python
dt.drop(['id'], axis=1, inplace=True)
```

```python
dt['person_home_ownership'] = dt['person_home_ownership'].map({'RENT': 0, 'MORTGAGE': 1, 'OWN':2, 'OTHER':3})
dt['loan_intent'] = dt['loan_intent'].map({'EDUCATION': 0, 'MEDICAL': 1, 'PERSONAL':2, 'VENTURE':3, 'DEBTCONSOLIDATION':4, 'HOMEIMPROVEMENT':5})
dt['loan_grade'] = dt['loan_grade'].map({'A': 0, 'B': 1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6})
dt['cb_person_default_on_file'] = dt['cb_person_default_on_file'].map({'N': 0, 'Y': 1})
dt.head()
```

```python
dt.info()
```

```python
sample_submission_url ='https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/sample_submission.csv?raw=true'

sf=pd.read_csv(sample_submission_url)
```

## 1. Train a KNN Classifier

```python
score_list = {}
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=22) #I've tried more than 50 values. 22 is the best value

KNN.fit(x_train,y_train)
knn_score = KNN.score(x_test,y_test)
print(f"Score is {knn_score}")
```

```python
# y_true = y_test
# y_pred = KNN.predict(x_test)
# plot_confusionMatrix(y_true,y_pred)
```

- Perform cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(KNN, x, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
score_list["KNN Classifier"] = scores.mean()
```

## 2. Train a Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(x_train,y_train)

lr_score = LR.score(x_test,y_test)


print(f"Score is {lr_score}")
```

- Perform cross validation

```python
scores = cross_val_score(LR, x, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
score_list["Logistic Regression"] = scores.mean()
```

## 3. Train a Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

nbc = GaussianNB()
nbc.fit(x_train,y_train)
nbc_score = nbc.score(x_test,y_test)

print(f"Score is {nbc_score}")
```

- Perform cross validation

```python
scores = cross_val_score(nbc, x, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
score_list["GaussianNBC"] = scores.mean()
```

## 4. Train a SVM Classifier

```python
svc = SVC()
svc.fit(x_train,y_train)
svc_score = svc.score(x_test,y_test)
print(f"Score is {svc_score}")
```

- Perform cross validation

```python
scores = cross_val_score(svc, x, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
score_list["SVC"] = scores.mean()
```

## 5. Train a Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1)
dtc.fit(x_train,y_train)

dtc_score = dtc.score(x_test,y_test)
print(f"Score is {dtc_score}")
```

- Perform cross validation

```python
scores = cross_val_score(dtc, x, y, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
score_list["DTC"] = scores.mean()
```

## 6. Train a Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=50,random_state=1)
rfc.fit(x_train,y_train)
rfc_score = rfc.score(x_test,y_test)
score_list["RFC"]=rfc_score

print(f"Score is {rfc_score}")
```

## 7. Compare all the performance of all classification models

```python
score_list = list(score_list.items())
```

```python
for alg,score in score_list:
    print(f"{alg} Score is {str(score)[:4]} ")
```

## 9. Generate Submission File

Choose the model that has the best performance to generate a submission file.

```python
id = sf.pop('id')
y_pred = rfc.predict(dt)

# Create a submission DataFrame
submission_df = pd.DataFrame({
    'id': id,
    'loan_status': y_pred
})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission_file.csv', index=False)
print("Submission file created: submission_file.csv")
```
